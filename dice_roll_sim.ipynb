```python
import pandas as pd
import random
import seaborn as sb
```

```python
def sim_dice_rolls(teams,rounds,num_rolls_per_round):
    ### create list that will hold data from all 40 teams
    class_list = []
    for i in range(0,teams):
        # create a list where each "team" puts their data
        team = []
        # each team completes 6 rounds total. Each round consists of taking 6 samples (aka rolling the dice 6 times) and computing the average value across all six rolls
        for j in range(0,rounds):
            # create a list to hold results from each sample
            one_sample = []
            for k in range(0,num_rolls_per_round):
                # simulate 6 dice rolls by selecting a random integer from 1-6 (inclusive) 6 times
                roll = random.randint(1,6)
                # add value of dice roll to your sample
                one_sample.append(roll)
            # take the average of your sample, this is your result for the round
            round_result = sum(one_sample)/len(one_sample)
            # add the value of your result for the round (aka the avg of your sample of 6 dice rolls) to your teams dataset
            team.append(round_result)
        class_list.append(team)
    
    ### make nested list into dataframe
    ## set index and column names
    # to have a nice Pandas table we set the index name (e.g. what each row should be named)
    index_names = ["Team " + str(i) for i in range(0,teams)]
    # we also set the names for each column
    col_names = ["Round " + str(i) for i in range(0,rounds)]
    
    # make dataframe
    # here we make and print a data frame
    dice_roll_sim = pd.DataFrame(data = class_list, index = pd.Series(index_names), columns=col_names)
    
    plot = sb.histplot(
    data = pd.melt(dice_roll_sim),
    x = 'value',
    stat = 'count',
    hue = 'variable',
    multiple = 'dodge',
    bins = 10
    )
    
    return dice_roll_sim,plot
```

```python
#Call function and show the dataframe 
#you can change the parameters here and play with them! 
my_df,my_plot = sim_dice_rolls(1000000,10,20)

my_df
```

```python
#show the sampling distribution
my_plot.figure
```

## Results
As we can see our sampling distribution we generated by conducting many random samples (e.g. dice throws) is roughly normally distributed. 

Compare this image to the one below which shows the population distribution for a dice roll. The population distribution for a dice roll is uniform (each value has an equal probability of showing up on a roll). It is clearly NOT a normal distribution.

Through this simulation we show that the central limit theorem does indeed work! We start with a population distribution that is completely not normal and we take random samples from this population, calculate some statistic about the sample, and plot the distribution of our sample statistic. This resulting distribution is called a *sampling distribution* and, if we did the sampling properly, our *sampling distribution* will be *normally* distributed.

The properties of a normal distribution allow us to make probabilistic claims with our sample statistic!
